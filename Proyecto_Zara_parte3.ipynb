```python
import pandas as pd 
```

```python
df_total = pd.read_csv("dftotal.csv")
```

```python
df_total.sample(30)
```

```python
df_total.dtypes
```

```python


df_total['Precio'] = df_total['Precio'].str.replace(' EUR', '', regex=False)  


df_total['Precio'] = df_total['Precio'].str.replace(',', '.', regex=False)  

# Convertir a float
df_total['Precio'] = df_total['Precio'].astype(float)  

```

```python
df_total.sample(20)
```

```python
precio_promedio = df_total['Precio'].mean()


if precio_promedio > 50:
    print("La hipótesis es verdadera: El precio promedio es superior a 50 euros.")
else:
    print("La hipótesis es falsa: El precio promedio no es superior a 50 euros.")


print(f"El precio promedio es: {precio_promedio:.2f} euros.")
```
